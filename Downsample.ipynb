```python
# Libraries
from google.cloud import bigquery
import pandas as pd
import numpy as np
import matplotlib as plt
# client location
client = bigquery.Client(location = "europe-north1")
print("Client creating using default project: {}".format(client.project))
```

N = 1

```python
# determine number of failures
query = """
    SELECT COUNT(failure)
    FROM df.m1days
    WHERE failure = 1
    """
query_job = client.query(query, location= "europe-north1")  # API request - starts the query
n_failure = query_job.to_dataframe()
print("N_failures:", n_failure.values[0][0])
```

```python
# determine number of healthy
query = """
    SELECT COUNT(failure)
    FROM df.m1days
    WHERE failure = 0
    """
query_job = client.query(query, location= "europe-north1")  # API request - starts the query
n_healthy = query_job.to_dataframe()
print("N_healthy:", n_healthy.values[0][0])
```

```python
# goal: 
to_drop = n_healthy.values[0][0] - n_failure.values[0][0]
to_drop
```

```python
# import dataset
query = """
    SELECT * EXCEPT(random)
    FROM df.M1N1D
    """
query_job = client.query(query, location= "europe-north1")  # API request - starts the query
M1N1D = query_job.to_dataframe()
M1N1D.head()
```

```python
M1N1D.groupby(["failure"]).agg(["count"])
```
